```python
class MyException(Exception):
    pass

print('+' * 30)
try:
    # sys.exit(100)
    # open('test')
    print('-' * 30)
    # raise MyException('我自定义异常')
    print('-' * 30)
    1/0

except MyException:
    print('My Exception')
except ZeroDivisionError:
    print('Zero')
except ArithmeticError:
    print('算术异常')
except OSError:  # if 如果有任何一个捕获了异常，其他异常捕获分支都不会执行
    print('OS Exception')
except Exception:
    print('异常+++')
except:    # 类似 else 一定放在最后 否则语法错误
    print('有异常')

# 异常类型 没有父子关系的，顺序随便
# 一旦有了父子类关系，子类在前(如果父类在前，后面的子类异常永远捕获不到)
# 如果有多个捕获： 从具体异常类型 到 宽泛的异常类型

print('+' * 30)  # 0
```

* 捕获规则
  - 捕获是从上到下一次比较 如果匹配 则执行匹配的except语句块
  - 如果被一个except语句捕获 其他except语句就不会再次捕获
  - 如果没有任何一个except语句捕获到这个异常，则该异常向外抛出
  - except: 称为缺省捕获，缺省捕获必须是最后一个捕获语句
  
* 捕获的原则
  - 从小到大 从具体到宽泛
  

```python
try ... except
try ... except 类型... except
try ... except 类型 ... except 类型 ... except 类型
try ... except 类型 ... except 类型 ... except ...
try ... finally

try ... except 类型 ... except 类型 ... except ... finally ... 
```

```python
try:
    f = None
    try:
        f = open('mysql.ini')
        print(f.read())
        print(f.closed, '---')
        f.write('abc')
    finally:
        if f:
            f.close()   # f 可能没有定义(风险) NameError
            print(f.closed, '+++')
except ArithmeticError:
    print('arith')
except FileNotFoundError:
    print('not found')
except OSError as e:
    print('os system error', e)
```
